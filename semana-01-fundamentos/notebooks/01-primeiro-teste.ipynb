## ðŸ“Š Tipos de GrÃ¡ficos com plt.subplot

### ðŸŽ¯ Guia Completo de VisualizaÃ§Ãµes para IA/ML

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Criar dados de exemplo para demonstrar diferentes tipos de grÃ¡ficos
np.random.seed(42)

# Dados simulados para exemplos
tempo = np.linspace(0, 10, 100)
vendas = 50 + 10 * tempo + np.random.normal(0, 5, 100)  # TendÃªncia linear com ruÃ­do
idades = np.random.normal(35, 10, 1000)  # DistribuiÃ§Ã£o normal de idades
experiencia = np.random.exponential(5, 1000)  # DistribuiÃ§Ã£o exponencial
salarios = 3000 + 200 * experiencia + np.random.normal(0, 500, 1000)

print("ðŸ“Š Dados criados para demonstraÃ§Ã£o:")
print(f"- Vendas ao longo do tempo: {len(vendas)} pontos")
print(f"- Idades dos funcionÃ¡rios: {len(idades)} amostras") 
print(f"- ExperiÃªncia vs SalÃ¡rio: {len(salarios)} observaÃ§Ãµes")
```

```python
# ðŸ“ˆ DEMONSTRAÃ‡ÃƒO: 4 Tipos de GrÃ¡ficos Essenciais para IA/ML

plt.figure(figsize=(15, 10))

# 1ï¸âƒ£ LINE PLOT - TendÃªncias temporais
plt.subplot(2, 2, 1)
plt.plot(tempo, vendas, 'b-', linewidth=2, label='Vendas')
plt.plot(tempo, 50 + 10 * tempo, 'r--', linewidth=2, label='TendÃªncia')
plt.title('ðŸ“ˆ Line Plot: Vendas ao Longo do Tempo')
plt.xlabel('Tempo (meses)')
plt.ylabel('Vendas (mil R$)')
plt.legend()
plt.grid(True, alpha=0.3)

# 2ï¸âƒ£ SCATTER PLOT - CorrelaÃ§Ãµes
plt.subplot(2, 2, 2)
colors = salarios  # Colorir por salÃ¡rio
scatter = plt.scatter(experiencia, salarios, c=colors, alpha=0.6, cmap='viridis')
plt.title('ðŸ“Š Scatter Plot: ExperiÃªncia vs SalÃ¡rio')
plt.xlabel('ExperiÃªncia (anos)')
plt.ylabel('SalÃ¡rio (R$)')
plt.colorbar(scatter, label='SalÃ¡rio')
plt.grid(True, alpha=0.3)

# 3ï¸âƒ£ HISTOGRAM - DistribuiÃ§Ãµes
plt.subplot(2, 2, 3)
plt.hist(idades, bins=30, alpha=0.7, color='green', edgecolor='black')
plt.title('ðŸ“Š Histogram: DistribuiÃ§Ã£o de Idades')
plt.xlabel('Idade (anos)')
plt.ylabel('FrequÃªncia')
plt.axvline(np.mean(idades), color='red', linestyle='--', linewidth=2, label=f'MÃ©dia: {np.mean(idades):.1f}')
plt.legend()
plt.grid(True, alpha=0.3)

# 4ï¸âƒ£ BOX PLOT - Outliers e quartis
plt.subplot(2, 2, 4)
data_boxplot = [idades, experiencia[:500], salarios[:500]/100]  # Normalizar salÃ¡rios
plt.boxplot(data_boxplot, labels=['Idades', 'ExperiÃªncia', 'SalÃ¡rios/100'])
plt.title('ðŸ“¦ Box Plot: ComparaÃ§Ã£o de DistribuiÃ§Ãµes')
plt.ylabel('Valores')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("ðŸŽ¯ USOS ESPECÃFICOS EM IA/ML:")
print("ðŸ“ˆ Line Plot: Monitorar loss/accuracy durante treinamento")
print("ðŸ“Š Scatter Plot: Visualizar clusters, correlaÃ§Ãµes entre features")  
print("ðŸ“Š Histogram: Verificar distribuiÃ§Ã£o de dados, detectar skewness")
print("ðŸ“¦ Box Plot: Identificar outliers, comparar grupos")
```

### ðŸ”§ **Tipos AvanÃ§ados para IA/ML**

#### **5. Heatmap (Matriz de CorrelaÃ§Ã£o)**
- **Uso:** Visualizar correlaÃ§Ãµes entre features
- **Quando:** Feature selection, anÃ¡lise exploratÃ³ria

#### **6. Violin Plot**
- **Uso:** DistribuiÃ§Ã£o + densidade dos dados
- **Quando:** Comparar distribuiÃ§Ãµes complexas

#### **7. Pair Plot**
- **Uso:** RelaÃ§Ãµes entre mÃºltiplas variÃ¡veis
- **Quando:** AnÃ¡lise exploratÃ³ria multivariada

#### **8. Confusion Matrix**
- **Uso:** Avaliar performance de classificaÃ§Ã£o
- **Quando:** ValidaÃ§Ã£o de modelos ML

```python
# ðŸ”¥ EXEMPLO AVANÃ‡ADO: Heatmap de CorrelaÃ§Ã£o
# Criar dataset correlacionado
data_corr = {
    'idade': np.random.normal(35, 10, 100),
    'experiencia': np.random.normal(8, 4, 100),
    'salario': [],
    'performance': []
}

# Criar correlaÃ§Ãµes intencionais
for i in range(100):
    # SalÃ¡rio correlacionado com idade e experiÃªncia
    salario = 3000 + 100 * data_corr['idade'][i] + 200 * data_corr['experiencia'][i] + np.random.normal(0, 500)
    data_corr['salario'].append(salario)
    
    # Performance correlacionada com experiÃªncia
    performance = 60 + 3 * data_corr['experiencia'][i] + np.random.normal(0, 10)
    data_corr['performance'].append(max(0, min(100, performance)))  # Limitar entre 0-100

df_corr = pd.DataFrame(data_corr)

# Calcular matriz de correlaÃ§Ã£o
correlation_matrix = df_corr.corr()

# Plotar heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
           square=True, linewidths=0.5)
plt.title('ðŸ”¥ Heatmap: Matriz de CorrelaÃ§Ã£o\n(Essencial para Feature Selection)')
plt.tight_layout()
plt.show()

print("ðŸ’¡ INTERPRETAÃ‡ÃƒO:")
print("ðŸ”´ Valores prÃ³ximos a +1: CorrelaÃ§Ã£o positiva forte")
print("ðŸ”µ Valores prÃ³ximos a -1: CorrelaÃ§Ã£o negativa forte") 
print("âšª Valores prÃ³ximos a 0: Sem correlaÃ§Ã£o")
print(f"\nðŸ“Š CorrelaÃ§Ãµes encontradas:")
print(f"Idade â†” SalÃ¡rio: {correlation_matrix.loc['idade', 'salario']:.3f}")
print(f"ExperiÃªncia â†” Performance: {correlation_matrix.loc['experiencia', 'performance']:.3f}")
```

## 4ï¸âƒ£ Conceitos Fundamentais IA/ML/DL

### ðŸ§  DefiniÃ§Ãµes BÃ¡sicas

**InteligÃªncia Artificial (IA)**
- Campo mais amplo
- MÃ¡quinas que simulam inteligÃªncia humana
- Inclui regras, lÃ³gica, ML, etc.

**Machine Learning (ML)**
- Subconjunto da IA
- Aprende padrÃµes nos dados
- Melhora performance com experiÃªncia

**Deep Learning (DL)**
- Subconjunto do ML
- Usa redes neurais profundas
- Inspirado no cÃ©rebro humano

```python
# Exemplo prÃ¡tico: correlaÃ§Ã£o simples (conceito ML)
correlacao = df['experiencia_anos'].corr(df['salario'])
print(f"ðŸ”— CorrelaÃ§Ã£o entre experiÃªncia e salÃ¡rio: {correlacao:.3f}")

if correlacao > 0.7:
    print("âœ… Forte correlaÃ§Ã£o positiva - mais experiÃªncia = maior salÃ¡rio")
elif correlacao > 0.3:
    print("âš ï¸ CorrelaÃ§Ã£o moderada")
else:
    print("âŒ CorrelaÃ§Ã£o fraca")

print("\nðŸ’¡ Este Ã© um exemplo bÃ¡sico de como ML encontra padrÃµes nos dados!")
```
