## 2Ô∏è‚É£ Primeiro Exemplo com Dados

```python
# Criar um dataset simples de exemplo
data = {
    'nome': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'idade': [25, 30, 35, 28, 32],
    'salario': [50000, 60000, 70000, 55000, 65000],
    'experiencia_anos': [2, 5, 8, 3, 6]
}

df = pd.DataFrame(data)
print("üìä Nosso primeiro dataset:")
print(df)
```

## 3Ô∏è‚É£ Primeira Visualiza√ß√£o

```python
# Configurar estilo dos gr√°ficos\n
plt.style.use('default')\n
plt.figure(figsize=(12, 4))\n
\n
# Gr√°fico 1: Rela√ß√£o experi√™ncia vs sal√°rio\n
plt.subplot(1, 2, 1)\n
plt.scatter(df['experiencia_anos'], df['salario'], color='blue', alpha=0.7, s=100)\n
plt.xlabel('Anos de Experi√™ncia')\n
plt.ylabel('Sal√°rio ($)')\n
plt.title('üìä Experi√™ncia vs Sal√°rio')\n
plt.grid(True, alpha=0.3)\n
\n
# Gr√°fico 2: Distribui√ß√£o de idades\n
plt.subplot(1, 2, 2)\n
plt.bar(df['nome'], df['idade'], color='green', alpha=0.7)\n
plt.xlabel('Pessoa')\n
plt.ylabel('Idade')\n
plt.title('üìä Distribui√ß√£o de Idades')\n
plt.xticks(rotation=45)\n
\n
plt.tight_layout()\n
plt.show()
```
