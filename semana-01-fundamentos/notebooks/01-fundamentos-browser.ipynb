## 2️⃣ Primeiro Exemplo com Dados

```python
# Criar um dataset simples de exemplo
data = {
    'nome': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'idade': [25, 30, 35, 28, 32],
    'salario': [50000, 60000, 70000, 55000, 65000],
    'experiencia_anos': [2, 5, 8, 3, 6]
}

df = pd.DataFrame(data)
print("📊 Nosso primeiro dataset:")
print(df)
```

```python
# Estatísticas básicas
print("📈 Estatísticas descritivas:")
print(df.describe())

print("\n🔍 Informações do dataset:")
print(df.info())
```

## 3️⃣ Primeira Visualização

```python
# Configurar estilo dos gráficos
import matplotlib.pyplot as plt
plt.style.use('default')
plt.figure(figsize=(12, 4))

# Gráfico 1: Relação experiência vs salário
plt.subplot(1, 2, 1)
plt.scatter(df['experiencia_anos'], df['salario'], color='blue', alpha=0.7, s=100)
plt.xlabel('Anos de Experiência')
plt.ylabel('Salário ($)')
plt.title('📊 Experiência vs Salário')
plt.grid(True, alpha=0.3)

# Gráfico 2: Distribuição de idades
plt.subplot(1, 2, 2)
plt.bar(df['nome'], df['idade'], color='green', alpha=0.7)
plt.xlabel('Pessoa')
plt.ylabel('Idade')
plt.title('📊 Distribuição de Idades')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## 4️⃣ Conceitos Fundamentais IA/ML/DL

### 🧠 Definições Básicas

**Inteligência Artificial (IA)**
- Campo mais amplo
- Máquinas que simulam inteligência humana
- Inclui regras, lógica, ML, etc.

**Machine Learning (ML)**
- Subconjunto da IA
- Aprende padrões nos dados
- Melhora performance com experiência

**Deep Learning (DL)**
- Subconjunto do ML
- Usa redes neurais profundas
- Inspirado no cérebro humano

```python
# Exemplo prático: correlação simples (conceito ML)
correlacao = df['experiencia_anos'].corr(df['salario'])
print(f"🔗 Correlação entre experiência e salário: {correlacao:.3f}")

if correlacao > 0.7:
    print("✅ Forte correlação positiva - mais experiência = maior salário")
elif correlacao > 0.3:
    print("⚠️ Correlação moderada")
else:
    print("❌ Correlação fraca")

print("\n💡 Este é um exemplo básico de como ML encontra padrões nos dados!")
```
