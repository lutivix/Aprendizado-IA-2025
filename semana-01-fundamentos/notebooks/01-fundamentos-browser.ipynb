## 2Ô∏è‚É£ Primeiro Exemplo com Dados

```python
# Criar um dataset simples de exemplo
data = {
    'nome': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'idade': [25, 30, 35, 28, 32],
    'salario': [50000, 60000, 70000, 55000, 65000],
    'experiencia_anos': [2, 5, 8, 3, 6]
}

df = pd.DataFrame(data)
print("üìä Nosso primeiro dataset:")
print(df)
```

```python
# Estat√≠sticas b√°sicas
print("üìà Estat√≠sticas descritivas:")
print(df.describe())

print("\nüîç Informa√ß√µes do dataset:")
print(df.info())
```

## 3Ô∏è‚É£ Primeira Visualiza√ß√£o

```python
# Configurar estilo dos gr√°ficos
import matplotlib.pyplot as plt
plt.style.use('default')
plt.figure(figsize=(12, 4))

# Gr√°fico 1: Rela√ß√£o experi√™ncia vs sal√°rio
plt.subplot(1, 2, 1)
plt.scatter(df['experiencia_anos'], df['salario'], color='blue', alpha=0.7, s=100)
plt.xlabel('Anos de Experi√™ncia')
plt.ylabel('Sal√°rio ($)')
plt.title('üìä Experi√™ncia vs Sal√°rio')
plt.grid(True, alpha=0.3)

# Gr√°fico 2: Distribui√ß√£o de idades
plt.subplot(1, 2, 2)
plt.bar(df['nome'], df['idade'], color='green', alpha=0.7)
plt.xlabel('Pessoa')
plt.ylabel('Idade')
plt.title('üìä Distribui√ß√£o de Idades')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## 4Ô∏è‚É£ Conceitos Fundamentais IA/ML/DL

### üß† Defini√ß√µes B√°sicas

**Intelig√™ncia Artificial (IA)**
- Campo mais amplo
- M√°quinas que simulam intelig√™ncia humana
- Inclui regras, l√≥gica, ML, etc.

**Machine Learning (ML)**
- Subconjunto da IA
- Aprende padr√µes nos dados
- Melhora performance com experi√™ncia

**Deep Learning (DL)**
- Subconjunto do ML
- Usa redes neurais profundas
- Inspirado no c√©rebro humano

```python
# Exemplo pr√°tico: correla√ß√£o simples (conceito ML)
correlacao = df['experiencia_anos'].corr(df['salario'])
print(f"üîó Correla√ß√£o entre experi√™ncia e sal√°rio: {correlacao:.3f}")

if correlacao > 0.7:
    print("‚úÖ Forte correla√ß√£o positiva - mais experi√™ncia = maior sal√°rio")
elif correlacao > 0.3:
    print("‚ö†Ô∏è Correla√ß√£o moderada")
else:
    print("‚ùå Correla√ß√£o fraca")

print("\nüí° Este √© um exemplo b√°sico de como ML encontra padr√µes nos dados!")
```
